```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Dhyey\Downloads\customer_shopping_behavior.csv")
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
df = df.rename(columns={ 'purchase_amount_(usd)':'purchase_amount'})
```

```python
# Create a column 'age_group' using quantile-based binning
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

```

```python

# Display the first 10 rows of age and age_group
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 14,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

```

```python
from sqlalchemy import create_engine
import urllib.parse  # Import this to handle the @ symbol

# Step 1: Connect to PostgreSQL
username = "postgres"
password = "S@xi1234"
# Encode the password to handle the '@' symbol
safe_password = urllib.parse.quote_plus(password)

host = "localhost"
port = "5432"
database = "customer_behavior"

# Create the SQLAlchemy engine using the safe_password
engine = create_engine(f"postgresql+psycopg2://{username}:{safe_password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "customer"
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
